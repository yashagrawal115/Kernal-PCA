```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A1_mean = [1, 1]
A1_cov =[[2, .99], [1, 1]]
A1 = np.random.multivariate_normal(A1_mean, A1_cov, 50)
plt.plot(A1[:,0], 'ro')
plt.plot(A1[:,1], 'bo')
```

```python
plt.hist(A1[:,0])
```

```python
plt.scatter(A1[:,0],A1[:,1])
```

```python
np.corrcoef(A1[:,0], A1[:,1])
```

```python
A2_mean = [5,5]
A2_cov =[[2,.99],[1,1]]
A2 = np.random.multivariate_normal(A1_mean, A1_cov, 50)
plt.plot(A2[:,0], 'ro')
plt.plot(A2[:,1], 'bo')
```

```python
plt.hist(A2[:,0])
```

```python
plt.scatter(A2[:,0],A2[:,1])
```

```python
np.corrcoef(A2[:,0], A2[:,1])
```

```python
A = np.vstack((A1, A2))
```

```python
A.shape
```

```python
plt.hist(A[:,1])
```

```python
B_mean = [5,0]
B_cov =[[.5,-1],[-.9,.5]]
B = np.random.multivariate_normal(B_mean, B_cov, 100)
```

```python
f = plt.figure(figsize=(10,10))

ax = f.add_subplot(111)
ax.set_title("a & b process")
ax.scatter(A[:,0], A[:,1], color ='r')
ax.scatter(B[:, 0], B[:,1], color='b')
```

```python
from sklearn import decomposition
```

```python
kpca = decomposition.KernelPCA(kernel = 'cosine', n_components=1)
```

```python
AB = np.vstack((A,B))
```

```python
AB_transformed = kpca.fit_transform(AB)
```

```python
A_color = np.array(['r']*len(B))
B_color = np.array(['b']*len(B))
colors = np.hstack((A_color, B_color))
```

```python
f = plt.figure(figsize = (10,4))
ax = f.add_subplot(111)
ax.set_title(" cosine kpca 1d process")

ax.scatter(AB_transformed, np.zeros_like(AB_transformed), color=colors)
```

```python
Y = np.zeros(200)
Y[100:] = 1
Y = Y.astype('int')
from sklearn.linear_model import LogisticRegression
LR_pca = LogisticRegression()
LR_pca.fit(AB, Y)
LR_pca.score(AB, Y)
```
